```python
import numpy as np

def check_outliers(X):
    q1 = np.percentile(X, 25)
    q3 = np.percentile(X, 75)
    IQR = q3 - q1
    
    threshold1 = q1 - 1.5 * IQR
    threshold2 = q3 + 1.5 * IQR

    count = 0
    L = []

    for i in X:
        if i < threshold1 or i > threshold2:
            count = count + 1
        else:
            L.append(i)

    print("No. of outliers: ", count)
    print("% of outliers: ", (count/len(X))/100)        
```

```python
check_outliers(df["Area"])
```

```python
check_outliers(df["Bedrooms"])
```

```python
check_outliers(df["Bathrooms"])
```

```python
check_outliers(df["Floors"])
```

```python
check_outliers(df["YearBuilt"])
```

```python
check_outliers(df["Price"])
```
