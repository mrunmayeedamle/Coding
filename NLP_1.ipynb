```python
from nltk.util import bigrams
text = "The cats were running faster than the dogs."
tokens = word_tokenize(text)
bigrams_list = list(bigrams(tokens))

for bigram in bigrams_list:
    print(bigram)
```

```python
words = len(gutenberg.words('milton-paradise.txt'))
print('No. of words =', words)
```

```python
char = len(gutenberg.raw('milton-paradise.txt'))
print('No. of characters =', char)
```

```python
sent = len(gutenberg.sents('milton-paradise.txt'))
print('No. of sentences =', sent)
```

```python
vocab = len(set(w for w in gutenberg.words('milton-paradise.txt'))) 
vocab
```

```python
avg_words = char/words
avg_sentence = words/sent
avg_vocab = sent/vocab

print('Average word count - ', avg_words)
print('Average sentence count - ', avg_sentence)
print('Average word per sentence count - ', avg_vocab)
```
